```python
new = df.filter(['overs_bowled', 'wickets_lost', 'run_rate', 'powerplay','cumulative_runs'], axis=1)
```

```python
ratio = 0.8
total_rows = new.shape[0]
train_size = int(total_rows*ratio)
```

```python
train = new[0:train_size]
test = new[train_size:]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
y=new['cumulative_runs']
x=new.drop(columns=['cumulative_runs'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
rf_model=RandomForestRegressor(n_estimators=200, max_depth=None, random_state=42,min_samples_leaf=1,min_samples_split=2)
```

```python
rf_model
```

```python
rf_model.fit(x_train,y_train)
pred=rf_model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,pred)
rf_r2=r2_score(y_test,pred)
```
