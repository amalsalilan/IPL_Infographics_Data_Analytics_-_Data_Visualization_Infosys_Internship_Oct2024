```python
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        plt.figure(figsize=(4, 2))
        df[col].hist(bins=30, edgecolor='black')
        plt.title(f'Histogram for {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()
```
