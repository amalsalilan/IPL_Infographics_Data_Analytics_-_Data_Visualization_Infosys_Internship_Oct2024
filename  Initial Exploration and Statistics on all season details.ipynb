```python
# 1. Distribution of runs scored per over
# Grouping by 'over' and summing the runs in each over
runs_per_over = data.groupby('over')['runs'].sum()
```

```python
runs_per_over
```

```python
# 2. Frequency of boundaries (4s and 6s)
# Filter dataset for boundaries and count each type (4s and 6s)
boundary_counts = data[data['isBoundary']].groupby('runs')['runs'].count()
```

```python
boundary_counts
```

```python
# 3. Dismissal Analysis: Count of dismissals and mean runs at dismissal
# Filter for rows with wicket information
dismissal_data = data.dropna(subset=['wicket_id'])
dismissal_count = dismissal_data.shape[0]
mean_runs_dismissal = dismissal_data['wkt_batsman_runs'].mean()
```

```python
dismissal_count
```

```python
 mean_runs_dismissal
```

```python
# 4. Top players based on batsman runs and bowler wickets
# Aggregating total runs for batsmen and total wickets for bowlers
top_batsmen = data.groupby('batsman1_name')['batsman1_runs'].sum().nlargest(5)
top_bowlers = data.groupby('bowler1_name')['bowler1_wkts'].sum().nlargest(5)
```
