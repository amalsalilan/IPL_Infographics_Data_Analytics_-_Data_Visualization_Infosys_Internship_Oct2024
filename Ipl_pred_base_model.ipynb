```python
import pandas as pd
import numpy as np
```

```python
first_dataset = pd.read_csv("/content/all_season_details.csv")
second_dataset = pd.read_csv('/content/all_season_summary.csv')
pd.set_option('display.max_columns', None)
first_dataset.head()
```

```python
first_dataset['season'].fillna(2023, inplace=True)
```

```python
total_runs = first_dataset.groupby(['match_id','innings_id'])['runs'].sum().reset_index()
```

```python
total_runs.head()
```

```python
total_runs = total_runs[total_runs['innings_id']==1]
```

```python
match_df = second_dataset.merge(total_runs[['match_id','runs']],left_on='id',right_on='match_id')
```

```python
match_df.head()
```

```python
match_df['home_team'].value_counts()
```

```python

teams = ['KKR',
         'CSK',
         'SRH',
         'RR',
         'MI',
         'RCB',
         'KXIP',
         'DC',
         'LSG',
         'GT'
         ]
```

```python
match_df = match_df[match_df['home_team'].isin(teams)]
match_df = match_df[match_df['away_team'].isin(teams)]
```

```python
match_df['home_team'].value_counts()
```

```python
match_df.shape
```

```python
match_df.columns
```

```python
match_df['decision'].value_counts()
```

```python
def add_col(row):
    if row['decision'] == 'BOWL FIRST':
        row['BowlingTeam'] = row['toss_won']
    else:
        if row['toss_won'] == row['home_team']:
            row['BowlingTeam'] = row['away_team']
        else:
            row['BowlingTeam'] = row['home_team']
    return row

# Apply the function row-wise and update the DataFrame
match_df = match_df.apply(add_col, axis=1)
```

```python
match_df.tail()
```

```python
match_df = match_df[['id','BowlingTeam','runs','winner']]
match_df.columns = ['id','BowlingTeam','total_runs_scored','winning_team']
```

```python
match_df.head()
```

```python
match_df.shape
```

```python
second_innings = match_df.merge(first_dataset,left_on='id',right_on='match_id')
second_innings = second_innings[second_innings['innings_id']==2]
second_innings.shape
```

```python
second_innings.head(10)
```

```python
second_innings['home_team'].value_counts()
```

```python
second_innings.columns
```

```python
def determine_batting_team(row):
  if row['BowlingTeam'] == row['home_team']:
    return row['away_team']
  else:
    return row['home_team']

second_innings['BattingTeam'] = second_innings.apply(determine_batting_team, axis=1)
```

```python
second_innings.head()
```

```python
second_innings['current_Score'] = second_innings.groupby('id')['runs'].cumsum()
```

```python
second_innings[['runs','current_Score']]
```

```python
second_innings['runs_left'] = second_innings['total_runs_scored'] - second_innings['current_Score']
```

```python
second_innings.head(10)
```

```python
second_innings['balls_left'] = 126 - (second_innings['over']*6 + second_innings['ball'])
second_innings
```

```python
second_innings.columns
```

```python
second_innings['player_out'] = second_innings['wkt_batsman_name'].fillna("0")
second_innings['player_out'] = second_innings['player_out'].apply(lambda x:x if x=="0" else "1")
second_innings['player_out'] = second_innings['player_out'].astype('int')
wickets = second_innings.groupby('id')['player_out'].cumsum()
second_innings['wickets_remaining'] = 10-wickets
second_innings
```

```python
second_innings['crr'] = round((second_innings['current_Score']*6)/(120 - second_innings['balls_left']),2)
second_innings['rrr'] = round((second_innings['runs_left']*6)/(second_innings['balls_left']),2)
second_innings
```

```python
def winner(team):
  return 1 if team['BattingTeam'] == team['winning_team'] else 0

second_innings['result'] = second_innings.apply(winner,axis=1)
second_innings
```

```python
required_data = second_innings[['BattingTeam','BowlingTeam','runs_left','balls_left','wickets_remaining','total_runs_scored','crr','rrr','result']].reset_index()
```

```python
required_data.head()
```

```python
required_data = required_data.replace([np.inf, -np.inf], np.nan)
required_data = required_data.dropna()
```

```python
columns_to_check = ['runs_left', 'balls_left', 'wickets_remaining', 'total_runs_scored', 'crr', 'rrr', 'result']
required_data = required_data.drop(['index'],axis=1)
required_data = required_data[~(required_data[columns_to_check] < 0).any(axis=1)]
```

```python
required_data.to_csv("out_data.csv",index=False)
```

```python
from sklearn.model_selection import train_test_split
x = required_data.iloc[:, :-1]
y = required_data.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['BattingTeam','BowlingTeam'])
])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[0]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42),
    'XGB Regressor': XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
}
```

```python
result = {}


for model_name, model in models.items():
    pipe = Pipeline([
        ('preprocessor', trf),
        ('regressor', model)
    ])

    pipe.fit(x_train, y_train)

    y_pred = pipe.predict(x_test)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    result[model_name] = {'MSE': mse, 'R2 Score': r2}

results_df = pd.DataFrame(result).T
print(results_df)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd

# Define the column transformer for categorical encoding
trf = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), ['BattingTeam', 'BowlingTeam'])
], remainder='passthrough')

# Define the models and their hyperparameter grids
models = {
    'Random Forest': (RandomForestRegressor(), {
        'model__n_estimators': [100, 200, 300],
        'model__max_depth': [10, 15, 20, None],
        'model__min_samples_split': [2, 5, 10],
    }),
}

# Dictionary to store results
result = {}

for model_name, (model, param_grid) in models.items():
    print(f"Tuning hyperparameters for {model_name}...")

    # Define the pipeline with encoding step and model
    pipe = Pipeline([
        ('transform', trf),
        ('model', model)
    ])

    # Check if hyperparameters need tuning
    if param_grid:
        search = RandomizedSearchCV(pipe, param_grid, scoring='r2', n_iter=10, cv=5, random_state=42)
        search.fit(x_train, y_train)
        best_model = search.best_estimator_
    else:
        best_model = pipe
        best_model.fit(x_train, y_train)

    y_pred = best_model.predict(x_test)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    result[model_name] = {'MSE': mse, 'R2 Score': r2}

print(pd.DataFrame(result).T)
```
