```python
# Check the percentage of missing values in each column
missing_values = ipl_df.isnull().sum() / len(ipl_df) * 100
missing_values[missing_values > 0].sort_values(ascending=False)
```

```python
# Display match_id for rows where 'season' is null
missing_season_matches = ipl_df[ipl_df['season'].isnull()]['match_id']

# Print match_id of rows with missing season values
print("Match IDs with missing season values:")
print(missing_season_matches.tolist())
```

```python
# Identify match_ids with missing season values that couldn't be filled
missing_season_match_ids = ipl_df[ipl_df['season'].isnull()]['match_id'].unique()
missing_season_match_ids
```

```python
# fill match_id 1359496, 1359538 with season 2023
ipl_df.loc[ipl_df['match_id'] == 1359496, 'season'] = 2023
ipl_df.loc[ipl_df['match_id'] == 1359538, 'season'] = 2023
```

```python
ipl_df['season'].isnull().sum()
```

```python
# Find duplicate rows based on all columns
dupes=ipl_df.duplicated().sum()
print("Before removing dupes",dupes)


# Remove duplicate rows
ipl_df = ipl_df.drop_duplicates()
dupes=ipl_df.duplicated().sum()
print("After removing dupes",dupes)
```

```python
#match between Virat Kohli as bats_man and Axar Patel as bowler
ipl_df.loc[(ipl_df['batsman1_name'] == 'Virat Kohli') & (ipl_df['bowler1_name'] == "Axar Patel")]
```

```python
# Group the data by 'season' and calculate the sum and mean of scores (or runs) for each season
grouped_df = ipl_df.groupby('season')['runs'].agg(['sum', 'mean'])

# Display the grouped DataFrame
grouped_df
```

```python
#match Count
match_count = ipl_df.groupby('season')['match_id'].count()
match_count
```

```python
#total score of each season
ipl_df['total_score'] = ipl_df.groupby('season')['runs'].transform('sum')
ipl_df.sample(20)
```
