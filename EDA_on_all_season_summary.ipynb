Importing required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Loading dataset

```python
df=pd.read_csv('downloads/all_season_summary.csv')
df.head(15)
```

Getting info about Columns

```python
df.columns
```

In analyzing the data for the "Most Number of Wins by Team," we can identify which teams have dominated the matches. This insight highlights the teams with the highest success rate and consistent performance across seasons. Here, MI as the most no of wins followed by CSK and KKR

```python
team_wins = df['winner'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=team_wins.index, y=team_wins.values, palette="viridis",hue=team_wins.index,legend=False)
plt.title('Most Number of Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of toss wins for each team
toss_wins = df['toss_won'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=toss_wins.index, y=toss_wins.values, palette="viridis",hue=toss_wins.index,legend=False)
plt.title('Most Number of Toss Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Toss Wins')
plt.xticks(rotation=45)
plt.show()
```

```python

# Count the number of tosses for each team using pd.concat
total_tosses = pd.concat([df['home_team'], df['away_team']]).value_counts()

# Count the number of toss wins for each team
toss_wins = df['toss_won'].value_counts()

# Calculate toss losses by subtracting toss wins from total tosses for each team
toss_losses = total_tosses - toss_wins

# Drop any NaN values that might appear for teams without toss losses recorded
toss_losses = toss_losses.dropna()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=toss_losses.index, y=toss_losses.values, palette="magma", hue=toss_losses.index,legend=False)
plt.title('Most Number of Toss Losses by Team')
plt.xlabel('Team')
plt.ylabel('Number of Toss Losses')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of matches played at each stadium
matches_per_stadium = df['venue_name'].value_counts()

# Plotting the horizontal bar chart
plt.figure(figsize=(12, 8))
sns.barplot(y=matches_per_stadium.index, x=matches_per_stadium.values, palette="Blues_d",hue=matches_per_stadium.index,legend=False)

# Adding data labels
for index, value in enumerate(matches_per_stadium.values):
    plt.text(value, index, str(value), va='center')

plt.title('Number of Matches Played at Each Stadium')
plt.xlabel('Number of Matches')
plt.ylabel('Stadium')
plt.show()
```

```python
# Count the number of matches with and without a Super Over
super_over_counts = df['super_over'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x=super_over_counts.index, y=super_over_counts.values, palette="pastel",hue=super_over_counts.index,legend=False)

# Setting title and labels
plt.title('Matches with and without a Super Over')
plt.xlabel('Super Over')
plt.ylabel('Number of Matches')

# Adding data labels
for index, value in enumerate(super_over_counts.values):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
# Count the number of times each player won "Player of the Match"
pom_counts = df['pom'].value_counts().head(10)  # Displaying top 10 players

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=pom_counts.values, y=pom_counts.index, palette="viridis")

# Adding titles and labels
plt.title('Top 10 Players with Most "Player of the Match" Awards')
plt.xlabel('Number of Awards')
plt.ylabel('Player')

plt.show()
```
