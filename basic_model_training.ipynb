```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('sample_dataset.csv')
data
```

```python
data.isnull().sum()
```

```python
new = data.filter(['overs_bowled', 'wickets_lost', 'run_rate', 'powerplay','cumulative_runs'], axis=1)
```

```python
new
```

```python
# Select split size
ratio = 0.8
total_rows = new.shape[0]
train_size = int(total_rows*ratio)
```

```python
train = new[0:train_size]
test = new[train_size:]
```

```python
train
```

```python
test
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
y=new['cumulative_runs']
x=new.drop(columns=['cumulative_runs'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
rf_model=RandomForestRegressor(n_estimators=200, max_depth=None, random_state=42,min_samples_leaf=1,min_samples_split=2)
```

```python
rf_model
```

```python
rf_model.fit(x_train,y_train)
pred=rf_model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,pred)
rf_r2=r2_score(y_test,pred)
```

```python
mse
```

```python
rf_r2
```

***K-Nearest Neighbors Regressor***

The k-nearest neighbors (KNN) regression model is a non-parametric machine learning algorithm that predicts the value of a new data point by using the values of its nearest neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_model = KNeighborsRegressor(n_neighbors=5)
```

```python
knn_model.fit(x_train, y_train)
```

```python
pred_knn = knn_model.predict(x_test)
```

```python
mse_knn = mean_squared_error(y_test, pred_knn)
r2_knn = r2_score(y_test, pred_knn)
```

```python
mse_knn
```

```python
r2_knn
```

***Decision Tree Regressor***

The Decision Tree model is simpler than Random Forest but can still capture non-linear relationships. Its useful to see if a single tree performs close to or better than an ensemble.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeRegressor(max_depth=None, random_state=42)
```

```python
dt_model.fit(x_train, y_train)
```

```python
pred_dt = dt_model.predict(x_test)
```

```python
mse_dt = mean_squared_error(y_test, pred_dt)
r2_dt = r2_score(y_test, pred_dt)
```

```python
mse_dt
```

```python
r2_dt
```

***Support Vector Regressor***

Support Vector Machines can be used for regression. They perform well on small to medium datasets and capture complex relationships.

```python
from sklearn.svm import SVR
```

```python
svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
```

```python
svr_model.fit(x_train, y_train)
```

```python
pred_svr = svr_model.predict(x_test)
```

```python
mse_svr = mean_squared_error(y_test, pred_svr)
r2_svr = r2_score(y_test, pred_svr)
```

```python
mse_svr
```

```python
r2_svr
```

***Randomsearch and grid search***

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
from scipy.stats import randint
```

```python
random_grid = {
    'n_estimators': randint(100, 500),
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 5)
}
```

```python
random_search = RandomizedSearchCV(estimator=rf_model, param_distributions=random_grid,
                                   n_iter=10, scoring='r2', cv=5, random_state=42, n_jobs=-1)
```

```python
random_search.fit(x_train, y_train)
```

```python
best_random_params = random_search.best_params_
```

```python
param_grid = {
    'n_estimators': [best_random_params['n_estimators'] - 50, best_random_params['n_estimators'], best_random_params['n_estimators'] + 50],
    'max_depth': [best_random_params['max_depth'] - 5, best_random_params['max_depth'], best_random_params['max_depth'] + 5] if best_random_params['max_depth'] else [None],
    'min_samples_split': [best_random_params['min_samples_split'] - 1, best_random_params['min_samples_split'], best_random_params['min_samples_split'] + 1],
    'min_samples_leaf': [best_random_params['min_samples_leaf'] - 1, best_random_params['min_samples_leaf'], best_random_params['min_samples_leaf'] + 1]
}
```

```python
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, scoring='r2', cv=5, n_jobs=2)
```

```python
print("Best Parameters from RandomizedSearchCV:", best_random_params)
```
