**Matches_per_season**

```python
matches_per_season=df.groupby('season')['match_id'].nunique().sort_values(ascending=False).head(10).reset_index()
matches_per_season
```

```python
palette=sb.set_palette("deep")
plt.figure(figsize=(10, 6))
bargraph=sb.barplot(x='season', y='match_id', data=matches_per_season, palette=palette,order=matches_per_season['season'],width=0.8)
for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_height()),
        (bar.get_x() + bar.get_width() / 2, bar.get_height()),
        ha='center',
        va='bottom' ,
        fontsize= 10,
        )

plt.title('Matches per season')
plt.xlabel('Season')
plt.ylabel('Matches')
plt.show()
```

**runs_per_season**

```python
runs_per_season=df.groupby(['season'])['runs'].sum().sort_values(ascending=False).head(10).reset_index()
runs_per_season
```

```python
palette=sb.set_palette("deep")
plt.figure(figsize=(10, 6))
bargraph=sb.barplot(x='season', y='runs', data=runs_per_season, palette=palette,order=matches_per_season['season'],width=0.8)
for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_height()),
        (bar.get_x() + bar.get_width() / 2, bar.get_height()),
        ha='center',
        va='bottom' ,
        fontsize= 10
    )

plt.title('runs per season')
plt.xlabel('Season')
plt.ylabel('runs')
plt.show()
```

**Runs by Team in Hometown**

```python
Hometeam_per_season=df.groupby('home_team')['runs'].sum().sort_values(ascending=False).head(10).reset_index()
Hometeam_per_season
```

```python
palette=sb.set_palette("deep")
plt.figure(figsize=(10, 6))
bargraph=sb.barplot(x='home_team', y='runs', data=Hometeam_per_season, palette=palette,order=Hometeam_per_season['home_team'],width=0.8)
for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_height()),
        (bar.get_x() + bar.get_width() / 2, bar.get_height()),
        ha='center',
        va='bottom' ,
        fontsize= 10
    )

plt.title('Home Teams by Total Runs Scored')
plt.xlabel('Season')
plt.ylabel('runs')
plt.show()
```

**Total Wides Bowled in Each IPL Season**

```python
wides_per_season=df.groupby('season')['isWide'].sum().sort_values(ascending=False).head(10).reset_index()
wides_per_season
```

```python
palette=sb.set_palette("deep")
plt.figure(figsize=(10, 6))
bargraph = sb.barplot(x='season', y='isWide', data=wides_per_season, order=wides_per_season['season'])
for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_height()),
        (bar.get_x() + bar.get_width() / 2, bar.get_height()),
        ha='center',
        va='bottom' ,
        fontsize= 10
    )

plt.title('wides per season')
plt.xlabel('Season')
plt.ylabel('runs')
plt.show()
```

**No balls per season**

```python
noballs_per_season = df.groupby(['season'])['isNoball'].sum().sort_values(ascending=False).reset_index()
print(noballs_per_season)
```

```python
plt.figure(figsize=(10, 6))

bargraph = sb.barplot(x='season', y='isNoball', data=noballs_per_season, order=noballs_per_season['season'])

for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_height()),
        (bar.get_x() + bar.get_width() / 2, bar.get_height()),
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.title('Total No-Balls Bowled in Each IPL Season', fontsize=16)
plt.xlabel('Season', fontsize=10)
plt.ylabel('Total No-Balls', fontsize=10)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Top 10 Away Teams**

```python
top_away_teams = df.groupby('away_team')['runs'].sum().sort_values(ascending=False).reset_index()
top_away_teams
```

```python
plt.figure(figsize=(10, 6))
bargraph = sb.barplot(x='runs', y='away_team', data=top_away_teams,)
for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_width()),
        (bar.get_width(), bar.get_y() + bar.get_height() / 2),
        ha='left',
        va='center',
        fontsize=9
    )
plt.title('Top 10 Away Teams by Total Runs Scored', fontsize=18)
plt.xlabel('Total Runs', fontsize=15)
plt.ylabel('Away Team', fontsize=15)
plt.tight_layout()
plt.show()
```

**Home team boundaries**

```python
home_team_boundaries = df.groupby('home_team')['isBoundary'].sum().sort_values(ascending=False).reset_index()
home_team_boundaries
```

```python
plt.figure(figsize=(10, 6))
bargraph = sb.barplot(x='home_team', y='isBoundary', data=home_team_boundaries)
for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_height()),
        (bar.get_x() + bar.get_width() / 2, bar.get_height()),
        ha='center',
        va='bottom',
        fontsize=9
    )
plt.title('Total Boundaries Scored by Each Home Team', fontsize=18)
plt.xlabel('Home_team', fontsize=15)
plt.ylabel('Total Boundaries', fontsize=15)
plt.tight_layout()
plt.show()
```
