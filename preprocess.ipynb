```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('winemag-data-130k-v2.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 0'])
```

```python
df.drop_duplicates()
```

```python
df['price'] = df['price'].fillna(df['price'].mean())
```

```python
df['points'] = df['points'].fillna(df['points'].mean())
```

```python
print(df.dtypes)
```

```python
df = pd.get_dummies(df, columns=['country', 'variety'], drop_first=True)
```

```python
scaler = MinMaxScaler()
df[['price', 'points']] = scaler.fit_transform(df[['price', 'points']])
```

```python
df.to_csv('preprocessed_wine_data.csv', index=False)
```
