```python
import pandas as pd
from google.colab import files
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Read the CSV file into a DataFrame
data = pd.read_csv("/content/sample_data/sample_dataset.csv")
copy_data=data.copy()
copy_data

```

```python
data.isnull().sum()
```

```python
# Define features and target variable
x = data[['overs_bowled', 'wickets_lost', 'run_rate', 'powerplay']]
y = data['cumulative_runs']
```

```python

# Split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Initialize and train the RandomForestClassifier
rf_model = RandomForestRegressor(n_estimators=200, max_depth=None, random_state=42)
```

```python

rf_model.fit(x_train, y_train)
```

```python
# Make a new prediction
new_data = pd.DataFrame({
    'overs_bowled': [10.2],
    'wickets_lost': [2],
    'run_rate': [7.5],
    'powerplay': [0]
})
predicted_category = rf_model.predict(new_data)
print(f"Predicted Cumulative Runs Category: {predicted_category[0]}")
```

```python
# Make predictions on the test set
y_pred = rf_model.predict(x_test)

print("Predictions on Test Set:", y_pred)
print("Actual Test Set Categories:", y_test.values)
```

```python

rf_simple_mse=mean_squared_error(y_test,y_pred)
rf_r2=r2_score(y_test,y_pred)
print("MSE",rf_simple_mse)
print("R2",rf_r2)
```
