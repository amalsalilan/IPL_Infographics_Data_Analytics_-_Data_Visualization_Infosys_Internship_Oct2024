1. Run Distribution per Season





```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/all_season_details.csv')


data['season'] = pd.to_numeric(data['season'], errors='coerce').fillna(0).astype(int)
data = data[data['season'] > 0]

# Calculate total runs per season
runs_per_season = data.groupby('season')['runs'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.bar(runs_per_season['season'], runs_per_season['runs'], color='skyblue')
plt.title('Total Runs Distribution per Season')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(runs_per_season['season'], rotation=45)
plt.tight_layout()
plt.show()
```

# Analyze the frequency of boundaries (isBoundary) per season and compare four/sixes across different teams.

```python


data['season'] = pd.to_numeric(data['season'], errors='coerce').fillna(0).astype(int)
data = data[data['season'] > 0]  # Exclude rows with season = 0


boundaries_per_season = data[data['isBoundary'] == 1].groupby('season').size()
sixes_per_season = data[data['runs'] == 6].groupby('season').size()


boundary_six_data = pd.DataFrame({
    'Boundaries': boundaries_per_season,
    'Sixes': sixes_per_season
}).fillna(0).reset_index()


plt.figure(figsize=(12, 6))
plt.bar(boundary_six_data['season'], boundary_six_data['Boundaries'], label='Boundaries', color='skyblue')
plt.bar(boundary_six_data['season'], boundary_six_data['Sixes'], bottom=boundary_six_data['Boundaries'], label='Sixes', color='orange')
plt.title('Boundaries and Sixes per Season')
plt.xlabel('Season')
plt.ylabel('Total Count')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(boundary_six_data['season'], rotation=45)
plt.tight_layout()
plt.show()
```

# **Average Runs per Over**

```python


data['season'] = pd.to_numeric(data['season'], errors='coerce').fillna(0).astype(int)
data = data[data['season'] > 0]


data['over_id'] = data['match_id'].astype(str) + "_" + data['innings_id'].astype(str) + "_" + data['over'].astype(str)
total_overs_per_season = data.groupby('season')['over_id'].nunique()
total_runs_per_season = data.groupby('season')['runs'].sum()


seasonal_stats = pd.DataFrame({
    'total_runs': total_runs_per_season,
    'total_overs': total_overs_per_season
})
seasonal_stats['avg_runs_per_over'] = seasonal_stats['total_runs'] / seasonal_stats['total_overs']


plt.figure(figsize=(12, 6))
plt.plot(seasonal_stats.index, seasonal_stats['avg_runs_per_over'], marker='o', color='b', linestyle='-', linewidth=2)
plt.title('Average Runs per Over per Season')
plt.xlabel('Season')
plt.ylabel('Average Runs per Over')
plt.grid(True)
plt.xticks(seasonal_stats.index, rotation=45)
plt.tight_layout()
plt.show()
```
