```python
import pandas as pd
import numpy as np
import matplotlib.pyplot   as plt
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
height=np.random.randint(140,200,100)
weight=np.random.randint(40,100,100)
df=pd.DataFrame({"Height":height,"Weight":weight})
df.head(5)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.hist(df['Height'],bins=15,color='skyblue',edgecolor='black')
plt.title('Height before scaling')
plt.xlabel("Height")
plt.ylabel("freq")

plt.subplot(1, 2, 2)
plt.hist(df['Weight'], bins=15, color='lightgreen', edgecolor='black')
plt.title('Weight Before Scaling')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df[['Height','Height']])
df_scaled=pd.DataFrame(scaled_data,columns=['Scaled Height', 'Scaled Weight'])
```

```python
# Plot histograms for the scaled height and weight data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df_scaled['Scaled Height'], bins=15, color='skyblue', edgecolor='black')
plt.title('Height After Scaling')
plt.xlabel('Scaled Height')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df_scaled['Scaled Weight'], bins=15, color='lightgreen', edgecolor='black')
plt.title('Weight After Scaling')
plt.xlabel('Scaled Weight')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Here are the **names of the symbols** used in the **StandardScaler** formula:

### Formula:
\[
\text{scaled\_value} = \frac{x - \mu}{\sigma}
\]

### Symbols and their names:

1. **\(x\)**:
   - **Name**: The **original value** (or individual data point).
   - This represents the actual value of the feature that you are standardizing.

2. **\(\mu\)**:
   - **Name**: **Mu** (Greek letter), which represents the **mean** (or average) of the feature.
   - This is the arithmetic mean of all values in the dataset for a particular feature.

3. **\(\sigma\)**:
   - **Name**: **Sigma** (Greek letter), which represents the **standard deviation**.
   - The standard deviation measures the spread or dispersion of the feature's values around the mean.

### Summary of Symbol Names:
- **\(x\)**: Original value or individual data point.
- **\(\mu\)**: Mu, the mean.
- **\(\sigma\)**: Sigma, the standard deviation.
