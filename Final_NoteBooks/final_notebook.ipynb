# 4. Modeling

## - Run Prediction Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression  
from sklearn.ensemble import GradientBoostingRegressor  
from xgboost import XGBRegressor 
```

```python
# Models to be used
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'XGB Regressor': XGBRegressor()
}
```

```python
# Load and preprocess data (assuming `df` is already preprocessed and ready)
# Select relevant features for prediction
features = [
    'over', 'ball', 'cumulative_runs', 'run_rate', 'wickets_lost', 
    'batsman1_runs', 'batsman1_balls', 'batsman2_runs', 'batsman2_balls', 
    'bowler1_overs', 'bowler1_maidens', 'bowler1_runs', 'bowler1_wkts', 
    'bowler2_overs', 'bowler2_maidens', 'bowler2_runs', 'bowler2_wkts'
] + [col for col in df.columns if col.startswith('home_team_') or col.startswith('away_team_') or col.startswith('current_innings_')]
```

```python
# Drop rows with missing values in the selected features
df = df.dropna(subset=features)
```

```python
# Define the features (X) and target (Y)
X = df[features]
Y = df['cumulative_runs']
```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
# Train and evaluate models
result = {}
for model_name, model in models.items():
    # Train the model
    model.fit(x_train, y_train)
    
    # Predict on test data
    y_pred = model.predict(x_test)
    
    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    # Store results
    result[model_name] = {'MSE': mse, 'R2 Score': r2}
```

```python
# Display Results
print("Model Performance Results >> \n")
for model, metrics in result.items():
    print(f"{model}: MSE={metrics['MSE']:.2f}, R2 Score={metrics['R2 Score']:.2f}\n")
```

# 6. Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Models to be used
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'XGB Regressor': XGBRegressor()
}
```

```python
param_grids = {
    'Linear Regression': {},  # No hyperparameters to tune for Linear Regression, it's a simple model.

    'Random Forest': {
        'n_estimators': [50, 100, 200],         # Number of trees in the forest. More trees can improve performance but increase computation time.
        'max_depth': [10, 20, None],            # Depth of each tree. Limiting depth can prevent overfitting, but too shallow may underfit.
        'min_samples_split': [2, 5, 10],        # Minimum samples required to split a node. Higher values can make the model more conservative.
        'min_samples_leaf': [1, 2, 4]           # Minimum samples required in each leaf node. Larger values help to smooth the model, but can increase bias.
    },

    'Gradient Boosting': {
        'n_estimators': [50, 100, 200],         # Number of boosting stages to be used. More stages generally improve performance at the cost of time.
        'learning_rate': [0.01, 0.1, 0.2],     # Step size at each iteration. Smaller values require more estimators but can lead to better performance.
        'max_depth': [3, 5, 7],                 # Maximum depth of the trees. Deeper trees can capture more complex relationships but can overfit.
    },

    'XGB Regressor': {
        'n_estimators': [50, 100, 200],         # Number of boosting rounds. Similar to `n_estimators` in Gradient Boosting.
        'learning_rate': [0.01, 0.1, 0.2],     # Step size for each boosting round. Smaller learning rates usually require more estimators to perform well.
        'max_depth': [3, 5, 7],                 # Maximum depth of the tree. Limits the complexity of the individual trees in the model.
        'subsample': [0.8, 1.0]                 # Fraction of samples used for training each tree. Lower values can prevent overfitting by adding randomness.
    }
}
```

```python
# Hyperparameter tuning function using RandomizedSearchCV
def tune_hyperparameters(model, param_grid, x_train, y_train, n_iter=10):
    """
    Tune hyperparameters for a given model using RandomizedSearchCV.

    Parameters:
    - model: The model to be tuned (e.g., RandomForestRegressor, XGBRegressor, etc.)
    - param_grid: Dictionary containing the hyperparameter grid
    - x_train: Training features
    - y_train: Training target variable
    - n_iter: Number of random combinations to sample

    Returns:
    - best_model: The model with the best found hyperparameters
    - best_params: The best hyperparameters found
    """
    randomized_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid,
                                           n_iter=n_iter, cv=3, n_jobs=4, verbose=2,
                                           scoring='neg_mean_squared_error', random_state=42)
    randomized_search.fit(x_train, y_train)
    
    # Return the best model and its hyperparameters
    best_model = randomized_search.best_estimator_
    best_params = randomized_search.best_params_
    
    return best_model, best_params
```

## - XGB Regressor Tuned

```python
# Define the best hyperparameters
best_hyperparameters = {
    'subsample': 0.8,
    'n_estimators': 200,
    'max_depth': 7,
    'learning_rate': 0.2
}
```

```python
# Initialize the XGBRegressor with the best hyperparameters
XGB_model = XGBRegressor(**best_hyperparameters)

# Train the model using the training data (x_train, y_train)
XGB_model.fit(x_train, y_train)

# Now, the model is trained and you can use it to make predictions on the test data
y_pred = XGB_model.predict(x_test)
```

# 7. Overfitting Check

```python
from sklearn.metrics import f1_score
```

```python
# Predict on training and testing data
X_train_pred = XGB_model.predict(x_train)
X_test_pred = XGB_model.predict(x_test)
```

```python
# Convert continuous predictions to discrete classes
def convert_to_classes(predictions, bins):
    return np.digitize(predictions, bins)  # Assign to bin
```

```python
# Define bins for classification (e.g., low, medium, high)
bins = [0, 10, 20, 30, 40, 50]  # Example bins based on your target's range
```

```python
# Convert both the actual values and predictions to classes
y_train_classes = convert_to_classes(y_train, bins)
y_test_classes = convert_to_classes(y_test, bins)
X_train_pred_classes = convert_to_classes(X_train_pred, bins)
X_test_pred_classes = convert_to_classes(X_test_pred, bins)

# Calculate F1 score for training and testing
train_f1 = f1_score(y_train_classes, X_train_pred_classes, average='weighted')
test_f1 = f1_score(y_test_classes, X_test_pred_classes, average='weighted')
```

```python
# Print results
print("Training F1 score:", train_f1)
print("Testing F1 score:", test_f1)

# Check for overfitting based on F1 score
if train_f1 > test_f1:
    print("Possible overfitting: Training performance is much better than testing.")
else:
    print("Model performance is balanced.")
```
