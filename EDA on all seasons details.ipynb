```python
import warnings
warnings.filterwarnings('ignore')

matches_per_season = df.groupby(['season'])['match_id'].nunique().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10, 6))
palette=sns.color_palette('viridis',n_colors=len(matches_per_season))

bargraph=sns.barplot(x='season', y='match_id', data=matches_per_season, palette=palette,order=matches_per_season['season'])

for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_height()), 
        (bar.get_x() + bar.get_width() / 2, bar.get_height()),  
        ha='center', 
        va='bottom' ,
        fontsize= 9
    )


plt.title('Number of Matches Played in Each IPL Season', fontsize=16)
plt.xlabel('Season', fontsize=10)
plt.ylabel('Number of Matches', fontsize=10)

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
warnings.filterwarnings('ignore')

total_runs_per_season = df.groupby(['season'])['runs'].sum().sort_values(ascending=False).reset_index()

print(total_runs_per_season)

plt.figure(figsize=(10, 6))

palette = sns.color_palette('viridis', n_colors=len(total_runs_per_season))
bargraph = sns.barplot(x='season', y='runs', data=total_runs_per_season,  order=total_runs_per_season['season'],palette=palette)

for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_height()), 
        (bar.get_x() + bar.get_width() / 2, bar.get_height()),  
        ha='center', 
        va='bottom',
        fontsize=9
    )

plt.title('Total Runs Scored in Each IPL Season', fontsize=16)
plt.xlabel('Season', fontsize=10)
plt.ylabel('Total Runs', fontsize=10)

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python

top_home_teams = df.groupby('home_team')['runs'].sum().sort_values(ascending=False).reset_index()

# Plotting the graph
plt.figure(figsize=(10, 6))
bargraph = sns.barplot(x='runs', y='home_team', data=top_home_teams, palette='viridis')

# Annotating the bars with values
for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_width()), 
        (bar.get_width(), bar.get_y() + bar.get_height() / 2),  
        ha='left', 
        va='center', 
        fontsize=9
    )

plt.title('Home Teams by Total Runs Scored', fontsize=16)
plt.xlabel('Total Runs', fontsize=14)
plt.ylabel('Home Team', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
top_away_teams = df.groupby('away_team')['runs'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
bargraph = sns.barplot(x='runs', y='away_team', data=top_away_teams, palette='plasma')

for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_width()), 
        (bar.get_width(), bar.get_y() + bar.get_height() / 2),  
        ha='left', 
        va='center', 
        fontsize=9
    )

plt.title('Top 10 Away Teams by Total Runs Scored', fontsize=16)
plt.xlabel('Total Runs', fontsize=14)
plt.ylabel('Away Team', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
home_team_runs = df.groupby('home_team')['runs'].sum()
away_team_runs = df.groupby('away_team')['runs'].sum()

total_team_runs = home_team_runs.add(away_team_runs)

total_team_runs = total_team_runs.sort_values(ascending=False).reset_index()
total_team_runs.columns = ['Team', 'Total Runs']

plt.figure(figsize=(12, 8))
bargraph = sns.barplot(x='Total Runs', y='Team', data=total_team_runs, palette='viridis')

for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_width()), 
        (bar.get_width(), bar.get_y() + bar.get_height() / 2),  
        ha='left', 
        va='center', 
        fontsize=9
    )

plt.title('Total Runs Scored by Each Team (Combined)', fontsize=16)
plt.xlabel('Total Runs', fontsize=14)
plt.ylabel('Team', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
home_team_wickets = df.groupby('home_team')['wicket_id'].count().sort_values(ascending=False).reset_index()
home_team_wickets.columns = ['Home Team', 'Total Wickets']

plt.figure(figsize=(12, 8))
bargraph = sns.barplot(x='Total Wickets', y='Home Team', data=home_team_wickets, palette='plasma')

for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_width()), 
        (bar.get_width(), bar.get_y() + bar.get_height() / 2),  
        ha='left', 
        va='center', 
        fontsize=9
    )

plt.title('Total Wickets Taken by Each Home Team', fontsize=16)
plt.xlabel('Total Wickets', fontsize=14)
plt.ylabel('Home Team', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
away_team_wickets = df.groupby('away_team')['wicket_id'].count().sort_values(ascending=False).reset_index()
away_team_wickets.columns = ['Away Team', 'Total Wickets']

plt.figure(figsize=(12, 8))
bargraph = sns.barplot(x='Total Wickets', y='Away Team', data=away_team_wickets, palette='viridis')

for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_width()), 
        (bar.get_width(), bar.get_y() + bar.get_height() / 2),  
        ha='left', 
        va='center', 
        fontsize=9
    )

plt.title('Total Wickets Taken by Each Away Team', fontsize=16)
plt.xlabel('Total Wickets', fontsize=14)
plt.ylabel('Away Team', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
home_team_wicks = df.groupby('home_team')['wicket_id'].count()

away_team_wicks = df.groupby('away_team')['wicket_id'].count()

total_team_wicks = home_team_wicks.add(away_team_wicks)

total_team_wicks = total_team_wicks.sort_values(ascending=False).reset_index()
total_team_wicks.columns = ['Team', 'Total Wickets']

plt.figure(figsize=(12, 8))
bargraph = sns.barplot(x='Total Wickets', y='Team', data=total_team_wicks, palette='viridis')

for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_width()), 
        (bar.get_width(), bar.get_y() + bar.get_height() / 2),  
        ha='left', 
        va='center', 
        fontsize=9
    )

plt.title('Total Wickets Taken by Each Team (Home and Away)', fontsize=16)
plt.xlabel('Total Wickets', fontsize=14)
plt.ylabel('Team', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
home_team_boundaries = df.groupby('home_team')['isBoundary'].sum()

total_boundaries_home = home_team_boundaries.sort_values(ascending=False).reset_index()
total_boundaries_home.columns = ['Home Team', 'Total Boundaries']

plt.figure(figsize=(12, 8))
bargraph = sns.barplot(x='Total Boundaries', y='Home Team', data=total_boundaries_home, palette='plasma')

for bar in bargraph.patches:
    bargraph.annotate(
        int(bar.get_width()), 
        (bar.get_width(), bar.get_y() + bar.get_height() / 2),  
        ha='left', 
        va='center', 
        fontsize=9
    )

plt.title('Total Boundaries Scored by Each Home Team', fontsize=16)
plt.xlabel('Total Boundaries', fontsize=14)
plt.ylabel('Home Team', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
away_team_boundaries = df.groupby('away_team')['isBoundary'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=away_team_boundaries.values, y=away_team_boundaries.index, palette='viridis')

for index, value in enumerate(away_team_boundaries.values):
    barplot.annotate(
        f'{value}', 
        (value, index), 
        ha='left', 
        va='center', 
        fontsize=9
    )

plt.title('Total Boundaries Scored by Away Teams', fontsize=16)
plt.xlabel('Total Boundaries', fontsize=14)
plt.ylabel('Away Team', fontsize=14)

plt.tight_layout()
plt.show()
```
