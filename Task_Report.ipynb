### 4.Extra Breakdowns

```python
Extra=df.groupby(['extras_type'])['extras_type'].count()
Extra
```

```python
Extra.plot(kind='pie', figsize=(5,5),title='Extras Distribution',autopct='%1.1f%%',startangle=90) 
```

#### Insights: The Total no.of Extra runs in the 20 overs because of different types

####    * From the various types of runs, wides were the most in scoring(59.3%) than the rest 
####    * There were no penalty(0.0%) in the matches
####    * 7.6% of the extras were no balls
####    * 28.3% were legbyes
####    * 4.8% were byes

```python
AvgRuns_Over=df.groupby('over')['total_runs'].mean()
AvgRuns_Over
```

```python
AvgRuns_Over_Graph = plt.plot(AvgRuns_Over)
plt.title('Average Runs per Over', ha='center')
plt.xlabel('Over')
plt.ylabel('Average Runs')
plt.xlim(0, 20)
plt.xticks([0, 5, 10, 15, 20])
plt.grid(True)
plt.show()
```

#### Insights: Average runs scored by a batsman in each over
####    * The average score in the first quarter of the match is 1.3 runs
####    * At half of the over the average score was 1.25 runs 
####    * On the 20th over the average score is nearly 1.8 runs
