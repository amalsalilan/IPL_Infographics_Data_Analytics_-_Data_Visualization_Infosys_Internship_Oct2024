**EDA ON THE DATASET**

```python
#Economy of Bowlers

# Grouping by 'bowler1_name' to calculate total runs conceded and overs bowled
bowler1_df = df.groupby('bowler1_name').agg(
    total_runs_conceded=('bowler1_runs', 'sum'),
    total_overs=('bowler1_overs', 'sum')
).reset_index()

# Calculating economy rate
bowler1_df['economy_rate'] = bowler1_df['total_runs_conceded'] / bowler1_df['total_overs']

# Filtering out invalid entries (total_overs > 0)
bowler1_df = bowler1_df[bowler1_df['total_overs'] > 0]

# Repeat for 'bowler2_name' if applicable
bowler2_df = df.groupby('bowler2_name').agg(
    total_runs_conceded=('bowler2_runs', 'sum'),
    total_overs=('bowler2_overs', 'sum')
).reset_index()

# Calculating economy rate for bowler2
bowler2_df['economy_rate'] = bowler2_df['total_runs_conceded'] / bowler2_df['total_overs']
bowler2_df = bowler2_df[bowler2_df['total_overs'] > 0]

# Combine both bowlers' data
combined_bowler_df = pd.concat([bowler1_df, bowler2_df], ignore_index=True)

# Removing duplicates and calculating final economy rates if needed
final_bowler_df = combined_bowler_df.groupby('bowler1_name', as_index=False).agg(
    total_runs_conceded=('total_runs_conceded', 'sum'),
    total_overs=('total_overs', 'sum')
)
final_bowler_df['economy_rate'] = final_bowler_df['total_runs_conceded'] / final_bowler_df['total_overs']

# Display the results
print("Bowlers' Economy Rates:")
print(final_bowler_df[['bowler1_name', 'economy_rate']].sort_values(by='economy_rate').head())
```

```python
#Batsman Strikerate
# Calculating strike rates for batsman1
batsman1_df = df.groupby('batsman1_name').agg(
    total_runs=('batsman1_runs', 'sum'),
    total_balls=('batsman1_balls', 'sum')
).reset_index()

# Calculating strike rate
batsman1_df['strike_rate'] = (batsman1_df['total_runs'] / batsman1_df['total_balls']) * 100

# Filtering out invalid entries (total_balls > 0)
batsman1_df = batsman1_df[batsman1_df['total_balls'] > 0]

# Repeat for 'batsman2_name' if applicable
batsman2_df = df.groupby('batsman2_name').agg(
    total_runs=('batsman2_runs', 'sum'),
    total_balls=('batsman2_balls', 'sum')
).reset_index()

# Calculating strike rate for batsman2
batsman2_df['strike_rate'] = (batsman2_df['total_runs'] / batsman2_df['total_balls']) * 100
batsman2_df = batsman2_df[batsman2_df['total_balls'] > 0]

# Combine both batsmen's data
combined_batsman_df = pd.concat([batsman1_df, batsman2_df], ignore_index=True)

# Removing duplicates and calculating final strike rates if needed
final_batsman_df = combined_batsman_df.groupby('batsman1_name', as_index=False).agg(
    total_runs=('total_runs', 'sum'),
    total_balls=('total_balls', 'sum')
)
final_batsman_df['strike_rate'] = (final_batsman_df['total_runs'] / final_batsman_df['total_balls']) * 100

# Display the results
print("Batsmen's Strike Rates:")
print(final_batsman_df[['batsman1_name', 'strike_rate']].sort_values(by='strike_rate', ascending=False).head())
```

```python
# Averages of batsman for each season
batsman_stats = df.groupby(['season', 'batsman1_name']).agg(
    total_runs=('batsman1_runs', 'sum'),
    dismissals=('wicket_id', 'count')  
).reset_index()

# Calculate the batting average
batsman_stats['batting_average'] = batsman_stats['total_runs'] / batsman_stats['dismissals']

# Replace NaN or infinity values with 0 (in case a batsman wasn't dismissed)
batsman_stats['batting_average'].fillna(0, inplace=True)

# Display the DataFrame with batsman averages
print(batsman_stats)

```

**MODELING**

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
# Select the required columns and create a copy
ball_by_ball_df = df[['season', 'match_id', 'home_team', 'away_team', 'current_innings',
                      'innings_id', 'over', 'ball', 'runs', 'wicket_id', 
                      'wkt_batsman_name', 'wkt_batsman_runs']].copy()

# Step 1: Calculate cumulative runs for each match and innings
ball_by_ball_df['cumulative_runs'] = ball_by_ball_df.groupby(['match_id', 'current_innings'])['runs'].cumsum()

# Step 2: Calculate run rate by dividing cumulative runs by the over (adding a small value to avoid division by zero)
ball_by_ball_df['run_rate'] = ball_by_ball_df['cumulative_runs'] / (ball_by_ball_df['over'] + 0.1)

# Step 3: Calculate wickets lost by counting instances where `wkt_batsman_runs` is non-negative
ball_by_ball_df['wickets_lost'] = ball_by_ball_df.groupby(['match_id', 'current_innings'])['wkt_batsman_runs']\
                                                 .transform(lambda x: (x >= 0).cumsum())

# Step 4: Fill any NaN values in the DataFrame with 0
ball_by_ball_df.fillna(0, inplace=True)
```

```python
X = ball_by_ball_df[['over', 'ball', 'run_rate','wickets_lost','season','match_id','home_team','away_team','current_innings']]
Y = ball_by_ball_df['cumulative_runs']
```

```python
cols = X.select_dtypes(include=['object']).columns.tolist()
```

```python
X = pd.get_dummies(X, columns=cols, drop_first=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Random forest regressor
rf_model=RandomForestRegressor(n_estimators=200, max_depth=None, random_state=42,min_samples_leaf=1,min_samples_split=2)
rf_model.fit(x_train,y_train)
y_rf_simple_pred=rf_model.predict(x_test)
rf_mse=mean_squared_error(y_test,y_rf_simple_pred)
rf_r2=r2_score(y_test,y_rf_simple_pred)
print("Random forest Regression:")
print(f"  Mean Squared Error: {rf_mse}")
print(f"  R-squared: {rf_r2}\n")
```

```python
#Linear regression
from sklearn.linear_model import LinearRegression


lr_model = LinearRegression()
lr_model.fit(x_train, y_train)

y_pred_lr = lr_model.predict(x_test)

mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression:")
print(f"  Mean Squared Error: {mse_lr}")
print(f"  R-squared: {r2_lr}\n")
```

```python
#decision tree regressor
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(x_train, y_train)

y_pred_dt = dt_model.predict(x_test)

mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print("Decision Tree Regressor:")
print(f"  Mean Squared Error: {mse_dt}")
print(f"  R-squared: {r2_dt}\n")
```

```python
#Gradient boost regressor
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(x_train, y_train)

y_pred_gb = gb_model.predict(x_test)

mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)

print("Gradient Boosting Regressor:")
print(f"  Mean Squared Error: {mse_gb}")
print(f"  R-squared: {r2_gb}\n")
```

```python
#Xgboost regressor
from xgboost import XGBRegressor


xgb_model = XGBRegressor(random_state=42)
xgb_model.fit(x_train, y_train)

y_pred_xgb = xgb_model.predict(x_test)

mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)

print("XGBoost Regressor:")
print(f"  Mean Squared Error: {mse_xgb}")
print(f"  R-squared: {r2_xgb}\n")
```

**Parameter Tuning**

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV



# Initialize the RandomForestRegressor
rf = RandomForestRegressor(random_state=42)

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],          # Number of trees in the forest
    'max_depth': [10, 20, None],             # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],         # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]            # Minimum number of samples required to be at a leaf node
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=2, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')

# Fit the model
grid_search.fit(x_train, y_train)

# Print the best parameters
print("Best parameters found: ", grid_search.best_params_)

# Evaluate on test data
best_model = grid_search.best_estimator_
test_score = best_model.score(x_test, y_test)
print("Test set score: ", test_score)
```
