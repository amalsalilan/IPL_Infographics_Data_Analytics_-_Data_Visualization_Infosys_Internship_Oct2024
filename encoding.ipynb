```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('store_df.csv')
data=df.copy()
df.head(3)
```

```python
ohe=OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
encoded_data=ohe.fit_transform(df[['Store_Type']])
df=pd.concat([encoded_data,df],axis=1).drop(columns=['Store_Type'])
```
