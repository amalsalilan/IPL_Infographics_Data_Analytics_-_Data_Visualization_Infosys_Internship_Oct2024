```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
```

```python
d1 = pd.read_csv('all_season_details.csv',low_memory=False, dtype={25: str})
df = d1.copy()
df
```

```python
retired_hurt = df[df["isRetiredHurt"]]

# Count the occurrences of players who retired hurt
player_retirement_counts = retired_hurt["batsman1_name"].value_counts()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    player_retirement_counts,
    labels=player_retirement_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.tab20.colors
)
plt.title("Distribution of Players Who Retired Hurt")
plt.show()
```

```python
# Filter the rows where batsman1_runs > 0 and ensure you're working on a copy
df_filtered = df[df["batsman1_runs"] > 0].copy()

# Replace zero balls with NaN in the filtered DataFrame to avoid division by zero
df_filtered["batsman1_balls"] = df_filtered["batsman1_balls"].replace(0, float('nan'))

# Calculate strike rate for batsman1
df_filtered["strike_rate_batsman1"] = (df_filtered["batsman1_runs"] / df_filtered["batsman1_balls"]).fillna(0) * 100

# Aggregate by season for batsman1
batting_trends = df_filtered.groupby("season").agg({
    "batsman1_runs": "sum",
    "batsman1_balls": "sum"
}).reset_index()

# Recalculate strike rate after aggregation
batting_trends["strike_rate_batsman1"] = (batting_trends["batsman1_runs"] / batting_trends["batsman1_balls"]).fillna(0) * 100

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=batting_trends, x="season", y="batsman1_runs", marker="o", label="Total Runs (Batsman 1)", color="blue")
sns.lineplot(data=batting_trends, x="season", y="strike_rate_batsman1", marker="o", label="Strike Rate (Batsman 1)", color="green")
plt.title("Batting Performance Trends Across Seasons (Batsman 1)", fontsize=16)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Performance Metrics", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset (assuming the data is already in a pandas DataFrame called 'df')
# df = pd.read_csv('path_to_data.csv')

# Set up a seaborn style
sns.set(style="whitegrid")

# 1. Distribution of runs scored across different seasons
plt.figure(figsize=(10,6))
sns.boxplot(x='season', y='runs', data=df)
plt.title('Distribution of Runs Across Seasons')
plt.show()

# 2. Number of wickets by each bowler in the dataset
# We need to count the wickets based on 'wkt_bowler_name'
wickets_by_bowler = df.groupby('wkt_bowler_name').size().reset_index(name='wickets')
plt.figure(figsize=(12,8))
sns.barplot(x='wickets', y='wkt_bowler_name', data=wickets_by_bowler.sort_values('wickets', ascending=False).head(15))
plt.title('Top 15 Bowlers by Number of Wickets')
plt.xlabel('Number of Wickets')
plt.ylabel('Bowler Name')
plt.show()

# 3. Strike rate comparison of batsmen (Batsman 1 vs Batsman 2)
plt.figure(figsize=(10,6))
sns.scatterplot(x='strike_rate_batsman1', y='batsman2_runs', data=df, alpha=0.6, hue='season')
plt.title('Strike Rate vs Runs for Batsman 2')
plt.xlabel('Strike Rate of Batsman 1')
plt.ylabel('Runs of Batsman 2')
plt.legend(title='Season', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 4. Heatmap of correlation between numerical features
num_cols = ['runs', 'strike_rate_batsman1', 'batsman1_runs', 'batsman1_balls', 'bowler1_runs', 'bowler1_overs', 'wkt_batsman_runs', 'wkt_batsman_balls']
correlation_matrix = df[num_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Key Numerical Features')
plt.show()

# 5. Distribution of runs scored by teams (Home vs Away)
plt.figure(figsize=(10,6))
sns.boxplot(x='home_team', y='runs', data=df)
plt.title('Runs Distribution by Home Team')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x='away_team', y='runs', data=df)
plt.title('Runs Distribution by Away Team')
plt.show()

# 6. Performance of each team (Runs scored by home and away teams)
team_performance = df.groupby(['home_team', 'away_team'])['runs'].sum().reset_index()
team_performance_sorted = team_performance.sort_values(by='runs', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x='runs', y='home_team', data=team_performance_sorted)
plt.title('Total Runs Scored by Each Home Team')
plt.show()

# 7. Identifying boundary frequency across overs
boundary_freq = df[df['isBoundary'] == True].groupby('over')['runs'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(x='over', y='runs', data=boundary_freq)
plt.title('Boundary Frequency Across Overs')
plt.xlabel('Over')
plt.ylabel('Runs Scored from Boundaries')
plt.show()

# 8. Visualizing the relationship between batsman strike rate and runs scored
plt.figure(figsize=(10,6))
sns.scatterplot(x='strike_rate_batsman1', y='batsman1_runs', data=df, hue='season', palette='coolwarm')
plt.title('Batsman 1 Strike Rate vs Runs Scored')
plt.xlabel('Strike Rate of Batsman 1')
plt.ylabel('Runs Scored by Batsman 1')
plt.show()

# 9. Checking for any significant relationship between retired hurt players and match performance
retired_hurt_data = df.groupby('isRetiredHurt')['runs'].sum().reset_index()

plt.figure(figsize=(8,6))
sns.barplot(x='isRetiredHurt', y='runs', data=retired_hurt_data)
plt.title('Runs Scored by Players who are Retired Hurt vs Not')
plt.show()

# 10. Plotting the distribution of overs bowled by the first bowler
plt.figure(figsize=(10,6))
sns.histplot(df['bowler1_overs'], kde=True, bins=20)
plt.title('Distribution of Overs Bowled by Bowler 1')
plt.show()

```

```python
import pandas as pd

# Assuming 'df' is your DataFrame containing the dataset

# Step 1: Filter out rows where batsman1_runs is 100 or more
centuries_df = df[df['batsman1_runs'] >= 100]

# Step 2: Group by season and batsman1_name, then count the number of centuries
centuries_count = centuries_df.groupby(['season', 'batsman1_name']).size().reset_index(name='centuries')

# Step 3: Find the batsman with the most centuries in each season
top_batsmen_by_season = centuries_count.loc[centuries_count.groupby('season')['centuries'].idxmax()]

# Display the results
print(top_batsmen_by_season[['season', 'batsman1_name', 'centuries']])
sns.set(style="whitegrid")

# Plot a barplot
plt.figure(figsize=(12,6))
sns.barplot(x='season', y='centuries', data=top_batsmen_by_season, hue='batsman1_name')

# Adding title and labels
plt.title('Batsman with Most Centuries in Each Season', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Number of Centuries', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

centuries_df = df[(df['batsman1_runs'] >= 100) | (df['batsman2_runs'] >= 100)]

batsman_data = pd.melt(centuries_df, id_vars=['season', 'match_id', 'match_name'],
                       value_vars=['batsman1_name', 'batsman2_name'],
                       var_name='batsman_position', value_name='batsman_name')

batsman_data['runs'] = pd.concat([centuries_df['batsman1_runs'], centuries_df['batsman2_runs']], axis=0).reset_index(drop=True)

centuries_data = batsman_data[batsman_data['runs'] >= 100]

centuries_count = centuries_data.groupby(['season', 'match_id', 'batsman_name']).size().reset_index(name='centuries')

top_batsmen_per_match = centuries_count.loc[centuries_count.groupby(['season', 'match_id'])['centuries'].idxmax()]

top_batsmen_per_match = top_batsmen_per_match.merge(df[['season', 'match_id', 'match_name']], on=['season', 'match_id'], how='left')

top_batsmen_per_season = top_batsmen_per_match.groupby('season')['centuries'].idxmax()
top_batsmen_per_season = top_batsmen_per_match.loc[top_batsmen_per_season]
print(top_batsmen_per_season)

plt.figure(figsize=(12, 6))

sns.barplot(data=top_batsmen_per_season, x='season', y='centuries', hue='batsman_name', dodge=False)

plt.title('Top Batsman with Most Centuries in Each Season', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Number of Centuries', fontsize=12)

plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# To suppress warnings, you can use this line (optional)
warnings.filterwarnings('ignore')

# Assuming 'df' is your DataFrame

# Step 1: Group by season, batsman, and match name, then sum the runs
df.loc[:, 'runs'] = df['batsman1_runs'] + df['batsman2_runs']  # Add runs from both batsmen if required
grouped = df.groupby(['season', 'batsman1_name', 'match_name'])['runs'].sum().reset_index()

# Step 2: Check if the total runs are greater than 100 and increment the count of centuries
grouped['century'] = grouped['runs'].apply(lambda x: 1 if x > 100 else 0)

# Step 3: Aggregate by season to count the total centuries per season
centuries_per_season = grouped.groupby('season')['century'].sum().reset_index()

# Step 4: Plot the data using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='season', y='century', data=centuries_per_season, palette='viridis')

# Set plot labels and title
plt.xlabel('Season')
plt.ylabel('Total Centuries')
plt.title('Total Centuries per Season')
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

df.loc[:, 'runs'] = df['batsman1_runs'] + df['batsman2_runs']
grouped = df.groupby(['season', 'batsman1_name', 'match_name'])['runs'].sum().reset_index()
grouped['century'] = grouped['runs'].apply(lambda x: 1 if x > 100 else 0)
centuries_per_batsman = grouped.groupby(['season', 'batsman1_name'])['century'].sum().reset_index()
top_batsman_per_season = centuries_per_batsman.loc[centuries_per_batsman.groupby('season')['century'].idxmax()]

plt.figure(figsize=(12, 6))
sns.barplot(x='season', y='century', data=top_batsman_per_season, hue='batsman1_name', palette='viridis')
plt.xlabel('Season')
plt.ylabel('Number of Centuries')
plt.title('Top Batsmen with Most Centuries in Each Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
