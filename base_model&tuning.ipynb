```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('processed_dataset.csv')
```

# Defining features and target

```python
X = df[[
    'over', 'ball', 'runs', 'isBoundary', 'isWide', 'isNoball',
    'batsman1_runs', 'batsman1_balls', 'bowler1_overs', 'bowler1_runs'
]]
Y = df['cumulative_score']
```

# Train-Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# Train the Random Forest model

```python
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, Y_train)
```

```python
Y_pred_rf = rf_model.predict(X_test)
```

# Calculate MSE and R² score

```python
mse_rf = mean_squared_error(Y_test, Y_pred_rf)
r2_rf = r2_score(Y_test, Y_pred_rf)
```

# Hyper parameter tuning

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import RandomizedSearchCV
```

```python
trf = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), ['isBoundary', 'isWide', 'isNoball'])
], remainder='passthrough')
```

```python
# Define models and hyperparameter grids
models = {
    'Random Forest Regressor': (RandomForestRegressor(), {
        'model__n_estimators': [100, 200, 300],
        'model__max_depth': [10, 15, 20, None],
        'model__min_samples_split': [2, 5, 10],
    }),
}

# Dictionary to store results
result = {}

# Loop through each model and perform hyperparameter tuning
for model_name, (model, param_grid) in models.items():
    print(f"Tuning hyperparameters for {model_name}...")

    # Define the pipeline with encoding step and model
    pipe = Pipeline([
        ('transform', trf),
        ('model', model)
    ])

    # Perform RandomizedSearchCV if there are hyperparameters to tune
    if param_grid:
        search = RandomizedSearchCV(pipe, param_grid, scoring='r2', n_iter=10, cv=5, random_state=42, n_jobs=-1)
        search.fit(X_train, Y_train)
        best_model = search.best_estimator_
    else:
        best_model = pipe
        best_model.fit(X_train, Y_train)

    # Make predictions and calculate metrics
    Y_pred = best_model.predict(X_test)
    mse = mean_squared_error(Y_test, Y_pred)
    r2 = r2_score(Y_test, Y_pred)

    # Store results
    result[model_name] = {
        'MSE': mse,
        'R2 Score': r2
    }

# Print results as a DataFrame
print(pd.DataFrame(result).T)
```
