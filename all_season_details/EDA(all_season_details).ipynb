```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
```

```python
d1 = pd.read_csv('all_season_details.csv',low_memory=False, dtype={25: str})
df = d1.copy()
df
```

# Top 10 Most Frequent Matches

```python
top_matches = df['match_name'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_matches.index, top_matches.values, color=sns.color_palette("viridis", len(top_matches)))
plt.title("Top 10 Most Frequent Matches")
plt.xlabel("Frequency")
plt.ylabel("Match Name")
plt.show()
```

# Runs Distribution Across Balls in a Match

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['runs'], bins=10, kde=True, color='blue')
plt.title("Runs Distribution Across Balls in a Match")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.show()
```

# Count of Boundaries and Non-Boundaries

```python
boundary_counts = df['isBoundary'].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=boundary_counts.index, y=boundary_counts.values, palette="coolwarm")
plt.title("Count of Boundaries and Non-Boundaries")
plt.xlabel("Is Boundary")
plt.ylabel("Count")
plt.xticks([0, 1], ['Non-Boundary', 'Boundary'])
plt.show()
```

# Count of Wides and No-balls

```python
plt.figure(figsize=(10, 4))

# Wides 
plt.subplot(1, 2, 1)
wide_counts = df['isWide'].value_counts()
sns.barplot(x=wide_counts.index, y=wide_counts.values, palette="Blues")
plt.title("Count of Wides")
plt.xticks([0, 1], ['Not Wide', 'Wide'])
plt.xlabel("Is Wide")
# +++++++++++++++++++++++++++++++++++++++++++++++++++
# No-balls 
plt.subplot(1, 2, 2)
noball_counts = df['isNoball'].value_counts()
sns.barplot(x=noball_counts.index, y=noball_counts.values, palette="Reds")
plt.title("Count of No-balls")
plt.xticks([0, 1], ['Not No-ball', 'No-ball'])
plt.xlabel("Is No-ball")
# +++++++++++++++++++++++++++++++++++++++++++++++++++
plt.tight_layout()
plt.show()
```

# Distribution of Overs Bowled by Bowlers

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['bowler1_overs'], bins=15, kde=True, color='purple')
plt.title("Distribution of Overs Bowled by Bowlers")
plt.xlabel("Overs Bowled")
plt.ylabel("Frequency")
plt.show()
```

# Total Runs by Batsman 'X'

```python
batsmanX_name='batsman1_name'
batsmanX_runs='batsman1_runs'

total_runs_batsman1 = df.groupby(batsmanX_name)[batsmanX_runs].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=total_runs_batsman1.index, x=total_runs_batsman1.values, palette="magma")
plt.title("Top 10 Total Runs by Batsman 1")
plt.xlabel("Total Runs")
plt.ylabel("Batsman 1 Name")
plt.show()
```

# Total Wickets by Bowler 'X'

```python
bowlerX_name = 'bowler1_name'  
bowlerX_wkts = 'bowler1_wkts' 

total_wkts_bowlerX = df.groupby(bowlerX_name)[bowlerX_wkts].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(y=total_wkts_bowlerX.index, x=total_wkts_bowlerX.values, palette="magma")
plt.title("Top 10 Bowlers by Total Wickets Taken")
plt.xlabel("Total Wickets")
plt.ylabel("Bowler Name")
plt.show()
```

# Comparison of Runs Scored by Batsmen

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['batsman1_runs', 'batsman2_runs']], palette="Set2")
plt.title("Comparison of Runs Scored by Batsman 1 and Batsman 2")
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.xticks([0, 1], ['Batsman 1', 'Batsman 2'])
plt.show()
```

# Innings-Wise Run Distribution

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='current_innings', y='runs', data=df, palette="cool")
plt.title("Innings-Wise Run Distribution")
plt.xlabel("Current Innings")
plt.ylabel("Runs")
plt.show()
```

# Match Results (Wins) Based on Season

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='season', palette="viridis", width=0.5)  
plt.title("Match Results Based on Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)  
plt.show()
```

# Distribution of Runs Scored in Each Match

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='runs', bins=30, kde=True, color='blue')
plt.title("Distribution of Runs Scored in Matches")
plt.xlabel("Runs Scored")
plt.ylabel("Frequency")
plt.show()
```

# Count of Boundaries Hit Over the Seasons

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='season', hue='isBoundary', palette="Set1")
plt.title("Count of Boundaries Hit Over Seasons")
plt.xlabel("Season")
plt.ylabel("Count of Boundaries")
plt.legend(title='Is Boundary', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

# Average Runs per Match by Home and Away Teams

```python
average_runs = df.groupby(['home_team'])['runs'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=average_runs, x='home_team', y='runs', palette="rocket")
plt.title("Average Runs Scored by Home Teams")
plt.xlabel("Home Team")
plt.ylabel("Average Runs")
plt.xticks(rotation=45)
plt.show()
```

# Most Successful Bowlers (by Wickets Taken)

```python
top_bowlers = df['bowler1_name'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette="viridis")
plt.title("Top 10 Bowlers by Number of Wickets")
plt.xlabel("Number of Wickets")
plt.ylabel("Bowlers")
plt.show()
```

# Relationship Between Batsman Runs and Balls Faced

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='batsman1_balls', y='batsman1_runs', alpha=0.6)
plt.title("Relationship Between Batsman Runs and Balls Faced")
plt.xlabel("Balls Faced by Batsman")
plt.ylabel("Runs Scored by Batsman")
plt.axhline(y=0, color='k', linestyle='--', lw=1)
plt.axvline(x=0, color='k', linestyle='--', lw=1)
plt.show()
```

# Total Runs Scored in Each Season

```python
total_runs_per_season = df.groupby('season')['runs'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=total_runs_per_season, x='season', y='runs', palette="pastel")
plt.title("Total Runs Scored in Each Season")
plt.xlabel("Season")
plt.ylabel("Total Runs")
plt.xticks(rotation=45)
plt.show()
```
