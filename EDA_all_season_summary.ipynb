```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_season_summary.csv')
```

```python
df.head()
```

```python
df.columns
```


# 1. MATCHES PER SEASON

TASK: Visualize the Total Matches per Season in IPL Data.

The line plot will display the number of matches per season, helping to identify increases, or decreases in the IPL match count over time.

```python
matches_per_season = df['season'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x=matches_per_season.index, y=matches_per_season.values, marker='o', color='b')
plt.title('Total Matches per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

# 2. TOP 5 TEAMS BY WINS

Task: Visualize the Top 5 Teams by Number of Wins in IPL Data.

From the plot, we observe that Which teams lead in the number of wins

```python
most_common_winners = df['winner'].value_counts().head(5)
plt.figure(figsize=(8, 6))
sns.barplot(x=most_common_winners.index, y=most_common_winners.values,color="pink")
plt.title('Top 5 Teams by Number of Wins')
plt.xlabel('Teams')
plt.ylabel('Wins')
plt.show()
```

# 3. Toss Decision Distribution

Task:Visualize the distribution of toss decisions (e.g., choosing to bat or bowl first) across matches.

```python
toss_decisions = df['decision'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(y=toss_decisions.index, x=toss_decisions.values,color="blue")
plt.title('Distribution of Toss Decisions')
plt.xlabel('Number of Matches')
plt.ylabel('Decision')
plt.show()
```

# 4. MOST COMMON VENUES

Task:Identify and visualize the top 5 venues where the most matches have been played.
This insight shows the most frequently used venues, revealing their popularity and significance in hosting matches.


```python
most_common_venues = df['venue_name'].value_counts().head(5)
plt.figure(figsize=(10, 6))
sns.barplot(x=most_common_venues.index, y=most_common_venues.values,color="green")
plt.title('Top 5 Most Common Venues')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

# 5. MOST NO.OF WINS BY EACH TEAM

Task: the teams with the highest number of wins.
This insight shows the teams with highest succesc rate and good performance accross the season

```python
most_wins = df['winner'].value_counts().head()
plt.figure(figsize=(10, 6))
sns.barplot(x=most_wins.index, y=most_wins.values)
plt.title('Top 5 Teams by Number of Wins')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.show()
```
