HANDILING MISSING VALUES

```python
df['season'] = df['season'].astype('object').fillna('Unknown')
df['batsman2_name'] = df['batsman2_name'].fillna('Unknown')
df['text']=df['text'].fillna('No Commentary')
```

```python
bowler2_columns = ['bowler2_id', 'bowler2_name', 'bowler2_overs', 'bowler2_maidens', 'bowler2_runs', 'bowler2_wkts']
df[bowler2_columns] = df[bowler2_columns].fillna('Unknown')
```

```python
wicket_columns = ['wicket_id', 'wkt_batsman_name', 'wkt_bowler_name', 'wkt_batsman_runs', 'wkt_batsman_balls', 'wkt_text']
df[wicket_columns] = df[wicket_columns].fillna('No Wicket') 
```

```python
df = df.drop(columns=['preText', 'postText']) 
```

```python
df.isnull().sum()
```

Feature Engineering

Tracks the total score i.e; cumulative runs for each team within each inning.


```python
df['cumulative_score'] = df.groupby(['match_id', 'current_innings'])['runs'].cumsum()
```

```python
df.cumulative_score
```

Outlier Removal

Removing outliers using IQR for 'runs'

```python
Q1_runs = df['runs'].quantile(0.25)
Q3_runs = df['runs'].quantile(0.75)
IQR_runs = Q3_runs - Q1_runs
df = df[(df['runs'] >= (Q1_runs - 1.5 * IQR_runs)) & (df['runs'] <= (Q3_runs + 1.5 * IQR_runs))]
```

Removing outliers using IQR for 'bowler1_overs'

```python
Q1_overs = df['bowler1_overs'].quantile(0.25)
Q3_overs = df['bowler1_overs'].quantile(0.75)
IQR_overs = Q3_overs - Q1_overs
df = df[(df['bowler1_overs'] >= (Q1_overs - 1.5 * IQR_overs)) & (df['bowler1_overs'] <= (Q3_overs + 1.5 * IQR_overs))]
```

```python
df.select_dtypes(include=['number'])
```

Standardization

Specify the numeric columns to standardize then Initialize the scaler and then Standardize the numeric columns

```python
from sklearn.preprocessing import StandardScaler
numeric_cols = ['runs', 'bowler1_overs', 'bowler1_runs', 'batsman1_runs']
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```
