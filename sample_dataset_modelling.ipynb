```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('sample_dataset.csv')
```

```python
df.head(5)
```

```python
X = df[['overs_bowled', 'wickets_lost', 'run_rate', 'powerplay']]
y = df['cumulative_runs']
```

```python
X
```

```python
y
```

```python
split=0.8
train_size=int((len(df))*split)
```

```python
data=df.sample(frac=1,random_state=42).reset_index(drop=True)
data
```

```python
train_data=data[:train_size]
test_data=data[train_size:]
```

```python
train_data
```

```python
test_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train ,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
rf_model = RandomForestRegressor(n_estimators = 200, max_depth=None, random_state=42 )
```

```python
rf_model
```

```python
rf_model.fit(x_train, y_train)
```

```python
y_pred=rf_model.predict(x_test)
y_pred
```

```python
rf_simple_mse=mean_squared_error(y_test,y_pred)
rf_r2=r2_score(y_test,y_pred)
```

```python
rf_simple_mse
```

```python
rf_r2
```
