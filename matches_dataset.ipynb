```python
#import the required libraries
import numpy as np
import pandas as pd
```

```python
#reading data into ipl
ipl=pd.read_csv('/content/matches.csv')
```

```python
#creating a copy of df
ipl_copy=ipl.copy()
ipl_copy
```

```python
#checking for null values or missing data
ipl_copy.isnull().sum()
```

```python
print(ipl_copy['city'].unique())
print(ipl_copy['city'].value_counts())
```

```python
#handling missing data of city by filling with Unknown in nan
ipl_copy.fillna({'city': 'Unknown'}, inplace=True)
ipl_copy.isnull().sum()
```

```python
# handling result_margin,target_runs,target_overs
ipl_copy['result_margin'] = ipl_copy['result_margin'].fillna(ipl_copy['result_margin'].median())
ipl_copy['target_runs'] = ipl_copy['target_runs'].fillna(ipl_copy['target_runs'].median())
ipl_copy['target_overs'] = ipl_copy['target_overs'].fillna(ipl_copy['target_overs'].median())
ipl_copy.isnull().sum()
```

```python
#since we dont know player_of_match and winner, we can remove the rows which have nan
ipl_copy.dropna(subset=['player_of_match', 'winner'], inplace=True)
#method column has 1074 missing values so dropping whole column
ipl_copy.drop(columns=['method'],inplace=True)
```
