```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
```

```python
data = {'Person': ['A', 'B', 'C'],
        'Height (cm)': [150, 180, 170],
        'Weight (kg)': [50, 70, 60]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
mean_height=df['Height (cm)'].mean()
mean_weight=df['Weight (kg)'].mean()
print(f"Mean Height:{mean_height} Mean weight:{mean_weight}")
```

```python
std_height=df['Height (cm)'].std()
std_weight=df['Weight (kg)'].std()
print(f"STD Height:{std_height} STD weight:{std_weight}")
```

```python
height_A=150
weight_A=50
```

```python
scaled_height_A=(height_A-mean_height)/std_height
print(scaled_height_A)
```

```python
scaled_wight_A=(weight_A-mean_weight)/std_weight
print(scaled_wight_A)
```

```python
# Now apply the same formula to the entire dataset
df['Scaled Height'] = (df['Height (cm)'] - mean_height) / std_height
df['Scaled Weight'] = (df['Weight (kg)'] - mean_weight) / std_weight

# Display the full DataFrame with scaled values
df[['Person', 'Scaled Height', 'Scaled Weight']]
```
