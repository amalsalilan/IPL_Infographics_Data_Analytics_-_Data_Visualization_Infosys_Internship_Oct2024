```python
import matplotlib.pyplot as plt
```

```python
winner_counts = ipl['winner'].value_counts()
plt.figure(figsize=(10, 6))
winner_counts.plot(kind='bar')
plt.title('Number of Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()
```

```python
wickets_per_bowler = ipl.groupby('bowler')['is_wicket'].sum()
wickets_per_bowler = wickets_per_bowler.sort_values(ascending=False)
wickets_per_bowler=wickets_per_bowler.head(10)
```

```python
plt.figure(figsize=(8, 5))
wickets_per_bowler.plot(kind='barh')
plt.show
```
