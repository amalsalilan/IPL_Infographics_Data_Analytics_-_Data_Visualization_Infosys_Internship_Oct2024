***2. DATA CLEANING***

```python
data_cleaned = data.drop(columns=['preText', 'postText','text'])
```

```python
data_cleaned['batsman2_id'] = pd.to_numeric(data_cleaned['batsman2_id'], errors='coerce')
data.head(5)
```

```python
data_cleaned.dropna(subset=['season', 'batsman2_name'], inplace=True)
data_cleaned.head(5)
```

```python
data_cleaned.fillna({
    'bowler2_id': 0, 'wicket_id': 0, 'text': '',
    'wkt_text': '', 'wkt_batsman_name': '',
    'wkt_bowler_name': '', 'wkt_batsman_runs': 0,
    'wkt_batsman_balls': 0
}, inplace=True)
```

```python
print(data_cleaned.duplicated().sum())
print(data_cleaned[data_cleaned.duplicated()])
```

```python
data_cleaned.drop_duplicates(inplace=True)
print(data_cleaned.duplicated().sum())
```

```python
data_cleaned.describe()
```

```python
pd.options.display.max_columns = None
```

```python
data_cleaned.head()
#all coulmns will be visible
```

***3. ANALYSIS of dataframe***

```python
data_cleaned['batsman1_name'].value_counts()
```

```python
print(data_cleaned['home_team'].unique())
print(data_cleaned['home_team'].nunique())
```

```python
data_cleaned.shape
```

```python
data_cleaned['season'].value_counts()
#this dataset only has 2023 season details
```

```python
# Group by 'season' and sum the 'runs'
runs_per_season = data_cleaned.groupby('season')['runs'].sum().reset_index()
print(runs_per_season.head())
print("only 2023 season is there")
```

```python
# Group by 'batsman1_name'
average_runs_per_batsman = data_cleaned.groupby('batsman1_name')['batsman1_runs'].mean().reset_index()
average_runs_per_batsman = average_runs_per_batsman.sort_values(by='batsman1_runs', ascending=False).head(10)
print(average_runs_per_batsman)
```

```python
# Group by 'bowler1_name'
wickets_per_bowler = data_cleaned.groupby('bowler1_name')['wicket_id'].count().reset_index()
wickets_per_bowler = wickets_per_bowler.sort_values(by='wicket_id', ascending=False).head(10)
print(wickets_per_bowler)
```

```python
# Group by 'batsman1_name' and sum the 'isBoundary' column
boundaries_per_batsman = data_cleaned.groupby('batsman1_name')['isBoundary'].sum().reset_index()
boundaries_per_batsman = boundaries_per_batsman.sort_values(by='isBoundary', ascending=False).head(10)
print(boundaries_per_batsman)
```

```python
# Group by 'away_team' and count 'wicket_id'
wickets_by_away_team = data_cleaned.groupby('away_team')['wicket_id'].count().reset_index()
wickets_by_away_team = wickets_by_away_team.sort_values(by='wicket_id', ascending=False).head(10)
print(wickets_by_away_team)
```

***4. VISUALIZATION***

```python
fig = px.histogram(data_cleaned, x='season', nbins=20, title="Distribution of Matches Across Seasons",template="simple_white")
fig.update_xaxes(title="Season")
fig.update_yaxes(title="Count of Matches")
fig.show()
```

```python
top_bowlers = data_cleaned['bowler1_name'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
top_bowlers.plot(kind='bar', color='orange')
plt.title("Top 10 Bowlers by Wickets Taken")
plt.xlabel("Bowler Name")
plt.ylabel("Number of Wickets")
plt.xticks(rotation=45)
plt.show()
```

```python
fig = px.pie(data_cleaned, names='isBoundary', title='Boundary vs Non-Boundary Balls', color_discrete_sequence=px.colors.qualitative.Pastel)
fig.show()
```

```python
top_batsmen = data_cleaned.groupby('batsman1_name')['batsman1_runs'].sum().nlargest(10)
fig = px.bar(top_batsmen, x=top_batsmen.index, y=top_batsmen.values, labels={'x': 'Batsman', 'y': 'Runs'},
             title="Top 10 Batsmen by Runs Scored")
fig.update_xaxes(title="Batsman")
fig.update_yaxes(title="Runs Scored")
fig.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['runs'], kde=True, bins=30, color='skyblue')
plt.title("Runs Distribution per Over")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.show()
```

```python
fig = px.scatter(data_cleaned, x='runs', y='wicket_id', color='current_innings', title="Runs vs Wickets by Innings",
                 labels={'runs': 'Runs', 'wicket_id': 'Wickets'})
fig.show()
```
